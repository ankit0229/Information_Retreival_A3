```python
import string
import re
import os
import copy
import statistics
import pickle
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter
lemmatizer = WordNetLemmatizer()

files = []
directory = r'D:\M.TECH SEM 2\IR\Assignments\A3\20_newsgroups\\'
for entry in os.listdir(directory):
    if os.path.isdir(os.path.join(directory, entry)):
        files.append(entry)

docs = []
directory = r'D:\M.TECH SEM 2\IR\Assignments\A3\20_newsgroups\\'
dict_counters = {}
dict_postings = {}
corpus_words = []
ctr = 1
all_doc_ids = []

list_folders_median_dfs = []
for fol in files:
    print(fol)
    temp_dir = os.path.join(directory, fol)
    docs = []
    dict_folder = {}
    folder_word_list = []
    complete_doc_loc = []
    for entry in os.listdir(temp_dir):
        if os.path.isfile(os.path.join(temp_dir, entry)):
            docs.append(entry)
            doc_loc = os.path.join(fol, entry)
            complete_doc_loc.append(doc_loc)

    all_doc_ids.extend(complete_doc_loc)
    for doc in docs:
        doc_loc = os.path.join(fol, doc)
        full_path = os.path.join(temp_dir, doc)
        fp = open(full_path, "r")
        text = fp.read()
        fp.close()
        ll = text.split("\n\n")
        del ll[0]
        text = "\n\n".join(ll)
        text = text.lower()
        text = re.sub(r'\S+@\S+', ' ', text)
        text = re.sub(r'[a-zA-Z]+[0-9]+', '', text)
        text = re.sub(r'[0-9]+[a-zA-Z]+', '', text)
        translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
        text = text.translate(translator)
        word_tokens = word_tokenize(text)
        stop_words = set(stopwords.words("english"))
        text = [word for word in word_tokens if word not in stop_words]
        lemmas = [lemmatizer.lemmatize(word) for word in text]
        list_words = copy.deepcopy(lemmas)
        dict_counts = Counter(list_words)
        dict_counters[doc_loc] = dict_counts
        dict_folder[doc_loc] = dict_counts
        unique_words = []
        for k in dict_counts:
            unique_words.append(k)
        corpus_words.extend(unique_words)
        folder_word_list.extend(unique_words)
#         for wd in unique_words:
#             if dict_postings.get(wd) is None:
#                 list_dc = [doc_loc]
#                 dict_postings[wd] = list_dc
#             else:
#                 dict_postings[wd].append(doc_loc)
    
    #Finding the median value of doc frequency of the folder
    dict_folder_df = {}
    for wds in folder_word_list:
        for kl in dict_folder:
            curr_dict = dict_folder[kl]
            if curr_dict[wds] != 0:
                if dict_folder_df.get(wds) is None:
                    dict_folder_df[wds] = 1
                else:
                    dict_folder_df[wds] = dict_folder_df[wds] + 1
    
    list_folder_dfs = []
    for ka in dict_folder_df:
        if dict_folder_df[ka] != 1 and dict_folder_df[ka] != 2 and dict_folder_df[ka] != 3 and dict_folder_df[ka] != 4:
            list_folder_dfs.append(dict_folder_df[ka])
    
    median_df = statistics.median(list_folder_dfs)
    list_folders_median_dfs.append(median_df)
                
print(len(dict_postings))

mean_df = statistics.median(list_folders_median_dfs)
print(f"The median df = {mean_df}")
```
